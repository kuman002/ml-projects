```python
df['total score'] = df['math score'] + df['reading score'] + df["writing score"]

df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading score']==100]['average'].count()
writing_full = df[df['writing score']==100]['average'].count()
math_full = df[df['math score']==100]['average'].count()

print("Number of students with full mark in Maths : ",math_full)
print(f'Number of students with full mark in Writing : {writing_full}')
print(f"Number of students with full mark in reading : {reading_full}")
```

```python
reading_less_20 = df[df['reading score']<=20]['average'].count()
writing_less_20 = df[df['writing score']<=20]['average'].count()
math_less_20 = df[df['math score']<=20]['average'].count()

print("Number of students with less than 20 marks in Maths : ",math_less_20)
print(f'Number of students with less than 20 marks in Writing : {writing_less_20}')
print(f"Number of students with less than 20 marks in reading : {reading_less_20}")
```

```python
gender_group = df.groupby('gender')['average'].mean()
gender_group
```

```python
# Ensure numeric columns
df['average'] = pd.to_numeric(df['average'], errors='coerce')
df['math score'] = pd.to_numeric(df['math score'], errors='coerce')

# Group by gender and take mean
gender_group = df.groupby('gender')[['average', 'math score']].mean().reset_index()

# Extract values safely
female_scores = gender_group.loc[gender_group['gender']=='female', ['average','math score']].values.flatten()
male_scores = gender_group.loc[gender_group['gender']=='male', ['average','math score']].values.flatten()

# X-axis labels
X = ['Total Average','Math Average']
X_axis = np.arange(len(X))

# Plot
plt.figure(figsize=(10,8))
plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male', color='skyblue')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female', color='pink')

# Labels and title
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average Marks by Gender", fontweight='bold')
plt.legend()
plt.show()
```
